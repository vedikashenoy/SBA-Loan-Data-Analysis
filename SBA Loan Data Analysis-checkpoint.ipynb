```python
#Create empty column
df['isDefault'] = ''
df.head()
```

```python
#Define a function
def isDefault(cols):
    ChgOffDate = cols[0]
    isDefault = cols[1]
    
    if pd.isnull(ChgOffDate):
        return 0
    else:
        return 1
```

```python
#Applythe function
df['isDefault'] = df[['ChgOffDate','isDefault']].apply(isDefault,axis=1)
```

```python
df['isDefault'] = pd.to_numeric(df['isDefault'])
type(df['isDefault'].iloc[0])
```

```python
#Read new file of clean data
df_clean = pd.read_csv('Loandatamodified.csv',low_memory=False)
```

```python
df_DT = df_clean[['LoanNr_ChkDgt','Term', 'NoEmp','NewExist', 'CreateJob', 'RetainedJob', 'FranchiseCode', 'UrbanRural','isDefault']]
```

```python
df_DT.head()
```

```python
df_DT = df_DT.dropna()
```

```python
#Split the data 
X = df_DT.drop('isDefault',axis=1)
y = df_DT['isDefault']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#Decision tree classification
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
#Random forest classification
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test,rfc_pred))
```

```python
print(classification_report(y_test,rfc_pred))
```
